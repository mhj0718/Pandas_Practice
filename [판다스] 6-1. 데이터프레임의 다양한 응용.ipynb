## 함수 매핑

```python
import seaborn as sns

# titanic 데이셋에서 age, fare 2개 열을 선택하여 df만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','fare']]
df['ten'] = 10
```

```python
df.head()
```

```python
# 사용자 함수 정의
def add_10(n):       # 10을 더하는 함수
    return n + 10
```

```python
add_10(10)
```

```python
def add_two_obj(a,b):
    return a + b
```

```python
add_two_obj(10,10)
```

```python
# 시리즈 객체에 적용
sr1 = df['age'].apply(add_10)         # n = df['age']의 모든 원소
sr1.head()
```

```python
# 람다 함수 활용: 시리즈 객체에 적용
sr3 = df['age'].apply(lambda x : add_10(x))  # x = df['age'], 반복문 대신 lambda 주로 사용
sr3.head()
```

```python
# 데이터프레임 전체에 applymap()으로 add_10() 함수를 매핑 적용
df_map = df.applymap(add_10)
df_map.head()

# 
```

```python
# axis = 0 >> 행 / axis = 1 >> 열  (기본적으로 그렇지만 함수별로 다르다)
# 사용자 함수 정의
def missing_value(series):      # 시리즈를 인수로 전달
    return series.isnull()      # 불린 시리즈를 반환

# 데이터프레임의 각 열을 인수로 전달하면 데이터프레임을 반환
result = df.apply(missing_value, axis=0)
result.head()
```

```python
# 사용자 함수 정의 
def min_max(x):
    return x.max() - x.min()

# 데이터프레임의 각 열을 인수로 전달하면 시리즈르 반환
result = df.apply(min_max)
result
```

```python
# 사용자 함수 정의
def add_two_obj(a,b):        # 두 객체의 합
    return a + b

# 데이터프레임의 2개 열을 선택하여 적용
# x=df
df['add'] = df.apply(lambda x : add_two_obj(x['age'], x['ten']), axis=1)
df.head()
```

```python
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','fare']]
# 각 열의 NaN 찾기 = 데이터프레임 전달하면 데이터프레임을 반환
def missing_value(x):
    return x.isnull()
```

```python
# 데이터프레임에 pipe() 메소드로 함수 매핑
result_df = df.pipe(missing_value)
result_df.head()
```

```python
# 각 열의 NaN 개수 반환 - 데이터프레임 전달하면 시리즈 반환
def missing_count(x):
    return missing_value(x).sum()
```

```python
df.isnull().sum()
```

```python
result_series = df.pipe(missing_count)
result_series
```

```python
# 데이터프레임의 총 NaN 개수 - 데이터프레임 전다라면 값을 반환
def total_number_missing(x):
    return missing_count(x).sum()
```

```python
result_value = df.pipe(total_number_missing)
result_value
```

## 데이터프레임 병합

```python
# 주식 데이터를 가져와서 데이터프레임 만들기
df1 = pd.read_excel('c:/py_data/stock price.xlsx', engine= 'openpyxl')
df2 = pd.read_excel('c:/py_data/stock valuation.xlsx', engine= 'openpyxl')
```

```python
df1.info()
```

```python
df2.info()
```

```python
# 데이터프레임 합치기 - 교집합
merge_inner = pd.merge(df1, df2, how='inner')
merge_inner
```

```python
# 데이터프레임 합치기 - 합집합
merge_outer = pd.merge(df1, df2, how='outer', on='id')
merge_outer
```

```python
# 데이터프레임 합치기 - 왼쪽 데이터프레임 기준, 키 값 분리
merge_left = pd.merge(df1, df2, how='left', left_on='stock_name',
                     right_on = 'name')

merge_left
```

```python
# 데이터프레임 합치기 - 오른쪽 데이터프레임 기준, 키 값 분리
merge_right = pd.merge(df1, df2, how='right', left_on='stock_name', 
                       right_on='name')
merge_right
```
