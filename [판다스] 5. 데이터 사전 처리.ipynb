## 범주형(카테고리) 데이터 처리

```python
# 구간분할
describe_dic = df_auto['출력'].describe()
describe_dic
```

```python
# [min, 50%, 75%, max]
bin_dividers = [describe_dic.loc['min'],describe_dic.loc['50%'],
                describe_dic.loc['75%'],describe_dic.loc['max']]
bin_dividers
```

```python
# 3개의 bin에 이름 지정
bin_names = ['저출력', '보통출력','고출력']
```

```python
# pd.cut 함수로 각 데이터를 3개의 bin에 할당
df_auto['출력등급'] = pd.cut(x=df_auto['출력'],        # 데이터 배열
                             bins=bin_dividers,        # 경계 값 리스트
                             labels=bin_names,         # bin 이름
                             include_lowest=True)      # 첫 경계값 포함
df_auto.head(30)
```

```python
# 원-핫 인코딩
# 출력등급 열의 범주형 데이터를 더미 변수로 변환
horsepower_dummies = pd.get_dummies(df_auto['출력등급'])
horsepower_dummies.head(30)
```

```python
# 원-핫 인코딩
# 토큰에 위치를 부여해서 입력값이 토큰값과 같으면 해당하는
# 위치에 지정해둔 값을 출력
# sklearn 라이브러리 불러오기
from sklearn import preprocessing
# 전처리를 위한 encoder 객체 생성
label_encoder = preprocessing.LabelEncoder()
onehot_encoder = preprocessing.OneHotEncoder()
```

```python
# label encoder로 문자열 범주를 숫자형 범주로 변환
onehot_labeled = label_encoder.fit_transform(df_auto['출력등급'].head(30))
onehot_labeled
# 0은 고출력, 1은 보통출력, 2는 저출력
```

```python
# 2차원 행렬로 형태 변경
onehot_reshaped = onehot_labeled.reshape(len(onehot_labeled),1)
onehot_reshaped
```

```python
# 희소행렬로 변환
onehot_fitted = onehot_encoder.fit_transform(onehot_reshaped)
print(onehot_fitted)
```
